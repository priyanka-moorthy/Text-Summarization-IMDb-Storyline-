```python
def getMoviesList():
  movies = []
  for idx in range(1,9):
    #change page number
    web1_url = "https://www.imdb.com/list/ls000634294/?sort=list_order,asc&st_dt=&mode=detail&page="+str(idx)+"&ref_=ttls_vm_dtl"

    web1_res = requests.get(web1_url)
    web1_sp = soup(web1_res.text, "html.parser")

    # extrat name url
    for i,tag in enumerate(web1_sp.findAll("h3", {"class": "lister-item-header"})):
      for j,res in enumerate(tag.findAll("a", href = True)):
        movies.append([res.text, res["href"]]) 
  return movies
```

```python
scrapedMovies = getMoviesList()
df = pd.DataFrame(scrapedMovies,
               columns =['Name', 'link'])
```

```python
def getMoviePlotSummary(lnk):
  baseUrl = "https://m.imdb.com"
  movieUrl = baseUrl+lnk

  movRes = requests.get(movieUrl)
  # print(movieUrl)
  movSoup = soup(movRes.text, "html.parser")
  summary = movSoup.find("div", {"class":"sc-16ede01-7"}).text
  return summary


# def getMovieStoryline(lnk):
#   baseUrl = "https://m.imdb.com"
#   movieUrl = baseUrl+lnk
#   print(movieUrl)
#   movRes = requests.get(movieUrl)
#   movSoup = soup(movRes.text, "html.parser")
#   storyline = movSoup.find("div",attrs={'data-testid':"storyline-plot-summary"})
#   print(storyline)
#   return storyline
# .find("div",{'class':'ipc-html-content-inner-div'})

```
